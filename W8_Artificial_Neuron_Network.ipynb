# INTRODUCTION ARTIFICIAL NEURON NETWORK

EXAMPLE 1

```python
from sklearn.linear_model import Perceptron
# Input data
x=[[0,0],[1,0],[0,1],[1,1]]
# Output data sample
y=[0,1,1,1]
# Model
model= Perceptron()
model.fit(x,y)
model.score(x,y)
```

```python
# New input data
x1=[[0.5,0],[0,0.5],[0.6,0],[0,0.6],[-10,2],[0.1,1],[0.3,4]]
# Result
y1=model.predict(x1)
print(y1)
```

# EXAMPLE 3

```python
from sklearn import datasets                         # Thư viện lấy dữ liệu
from sklearn.linear_model import Perceptron          # Lấy model preceptron
from sklearn.model_selection import train_test_split # Lấy hàm train test split
from sklearn.metrics import accuracy_score           # Đánh giá sai số nằm trong thư mục metrics -> lấy hàm accuracy_score
from sklearn.preprocessing import StandardScaler     # Chuẩn hóa phân vùng dữ liệu
```

```python
iris =datasets.load_iris() # lấy dữ liệu iris
X=iris.data
y=iris.target
print(X[:5])
print(y[:5])
```

```python
# chia tỉ lệ train và test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3) # tỉ lệ 70/30
```

```python
#Scale dữ liệu
sc=StandardScaler()
sc.fit(X_train)
x_train_std=sc.transform(X_train)
x_test_std =sc.transform(X_test)
print(X_train)
print(x_train_std)
```

STUDY WITHOUT SCALE DATA

```python
# max_iter số lần học
# eta0 tốc độ học
# random_state lấy ngẫu nhiên
model = Perceptron(max_iter = 1000, eta0 = 1, random_state = 0)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print(y_predict)
print(y_test)
print('Độ chính xác mô hình chưa scale: ', accuracy_score(y_test, y_predict))

model = Perceptron(max_iter = 1000, eta0 = 1, random_state = 0)
model.fit(x_train_std, y_train)
y_predict = model.predict(x_test_std)
print(y_predict)
print(y_test)
print('Độ chính xác mô hình: ', accuracy_score(y_test,y_predict))
```

# crabs.csv

```python
import pandas as pd
url ='crabs.csv'
data = pd.read_csv(url)
data.head()
```

```python
data['sex']=data['sex'].replace(["F", "M"], ["0", "1"])
data['speies']=data['speies'].replace(["O", "B"], ["0", "1"])
data
```

```python
X=data.loc[:,['speies','index','fontallip','rearwidth','length','width','depth']]
y=data['sex']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
#Scale dữ liệu
sc=StandardScaler()
sc.fit(X_train)
x_train_std=sc.transform(X_train)
x_test_std =sc.transform(X_test)
print(X_train)
print(x_train_std)
```

```python
# max_iter số lần học
# eta0 tốc độ học
# random_state lấy ngẫu nhiên
model = Perceptron(max_iter = 1000, eta0 = 1, random_state = 0)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
print(y_predict)
print(y_test)
print('Độ chính xác mô hình chưa scale: ', accuracy_score(y_test, y_predict))

model = Perceptron(max_iter = 1000, eta0 = 1, random_state = 0)
model.fit(x_train_std, y_train)
y_predict = model.predict(x_test_std)
print(y_predict)
print(y_test)
print('Độ chính xác mô hình: ', accuracy_score(y_test,y_predict))
```
