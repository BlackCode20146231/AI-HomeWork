```python
from PIL import Image
import pytesseract
import numpy as np
import matplotlib.pyplot as plt
file='book.jpg'
img=plt.imread(file)
plt.imshow(img)
plt.show()
img1=np.array(Image.open(file))
text=pytesseract.image_to_string(img1)
print(text)
```

```python
from pytesseract import Output
import pytesseract 
import cv2 as cv
result=pytesseract.image_to_data(img1,output_type = Output.DICT)
print(result)
for i in range(0,len(result['text'])):
  x = result['left'][i]
  y = result['top'][i]
  w = result['width'][i]
  h = result['height'][i]
  conf=int(result['conf'][i])
  if conf > 70:
    text=''.join([c if ord(c)<128 else '' for c in text]).strip()
    cv.rectangle(img1,(x,y),(x+w,y+h),(0,255,0),2)
    cv.putText(img1,text,(x,y-10),cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,200),2)
plt.imshow(img1)
```

```python
from transformers import pipeline
classifier=pipeline('zero-shot-classification',
                    model='roberta-large-mnli')
text='I get up early and prepare the breakfast'
labels=['travel','cooking','bread','morning']
classifier(text, labels)
```

```python
from transformers import pipeline

classifier = pipeline('zero-shot-classification', model='roberta-large-mnli')

text = 'The final exam is tomorrow, what should I do?'
labels = ['Sleep', 'Read a book', 'Do homework', 'Nothing']

result = classifier(text, labels)

print(f"Text: {text}\nLabels: {result['labels']}\nScores: {result['scores']}")
```
